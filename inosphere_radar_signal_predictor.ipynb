```python
import pandas as pd
url="https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/Ionosphere.csv"
df=pd.read_csv(url)
df.head()
```

```python
x=df.drop(["Class"],axis=1)
y=df["Class"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from keras.models import Sequential, load_model
from keras.layers import Dense,Dropout
```

Build model

```python
model=Sequential()
model.add(Dense(units=10,activation='relu',input_dim=len(x_train.columns)))
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32)
```

```python
model.save('weight.keras')
```

```python
from keras.models import load_model

# Load the saved model
loaded_model = load_model('weight.h5')

# Display model summary to confirm it's loaded correctly
loaded_model.summary()
```

```python
# Make predictions on the test set
predictions_proba = loaded_model.predict(x_test)

# Display the first 10 probability predictions
print("First 10 probability predictions:")
print(predictions_proba[:10])

# Convert probabilities to binary class predictions (0 or 1)
# A common threshold is 0.5
predictions_binary = (predictions_proba > 0.5).astype(int)

# Display the first 10 binary predictions
print("\nFirst 10 binary predictions:")
print(predictions_binary[:10])
```

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy of the predictions
accuracy = accuracy_score(y_test, predictions_binary)
print(f"Test Accuracy: {accuracy:.4f}")
```
